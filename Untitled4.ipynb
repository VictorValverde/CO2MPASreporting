```python
Inputfile = r'D:\co2mpas-version-trials\20160610\JELICA files VOLVO\co2mpas_template_S90_test1.xlsx'
Outputfile = r'D:\co2mpas-version-trials\20160603\out_jelica\20160614_152710-co2mpas_template_S90_test1.xlsx'
```

```python
WLTPHi = pd.read_excel(Inputfile,'WLTP-H', header = [0,1], index_col=[0])
WLTPHi.columns = WLTPHi.columns.droplevel()
WLTPHo = pd.read_excel(Outputfile,'wltp_h_predictions_time_series', header = [0,1], index_col=[0])
WLTPHo.columns = WLTPHo.columns.droplevel()
```

```python
def PlotSeries(series=['fuel_consumptions', 'engine_speeds_out','engine_powers_out', 'velocities',
                              'engine_coolant_temperatures','state_of_charges','battery_currents','gear_box_torque_losses',
                             'alternator_currents','gear_box_temperatures']):
    f, axarr = plt.subplots(len(series),1, sharex=True, figsize=(12,46))
    for index,item in enumerate(series):
        try:
            axarr[index].plot(WLTPHi[item], label = 'target', color='red')
        except:
            pass
        try:
            axarr[index].plot(WLTPHo[item], label = 'prediction', color='blue')
        except:
            pass
        axarr[index].grid(True, color='grey', linestyle='-')
        axarr[index].set_title(item, fontsize=20)
        axarr[index].legend(loc='upper left', bbox_to_anchor=(-0.008, 1.018),fancybox=True, shadow=True, ncol=4)
        plt.setp(axarr[index].get_xticklabels(), fontsize=10, visible=True)
        plt.subplots_adjust(hspace = .2)
    plt.subplots_adjust(bottom=0.02, right=0.9, top=0.98)
    plt.show()
```

```python
PlotSeries()
```
